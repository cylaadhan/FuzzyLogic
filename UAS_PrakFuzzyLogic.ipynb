```python
def mu_suhu_normal(x):
    if x <= 36:
        return 1.0
    elif 36 < x < 37.5:
        return (37.5 - x) / (37.5 - 36)
    else:
        return 0.0

def mu_suhu_demam(x):
    if x <= 37:
        return 0.0
    elif 37 < x < 39:
        return (x - 37) / (39 - 37)
    else:
        return 0.0

def mu_suhu_tinggi(x):
    if x <= 38:
        return 0.0
    elif 38 < x < 41:
        return (x - 38) / (41 - 38)
    else:
        return 1.0

def trombosit_low_normal(x, batas=150):
    return (1.0, 0.0) if x <= batas else (0.0, 1.0)

def leukosit_low_normal(x, batas=4.0):
    return (1.0, 0.0) if x <= batas else (0.0, 1.0)

suhu = 38.5
trombosit = 120
leukosit = 3.8

mu_norm   = mu_suhu_normal(suhu)
mu_demam  = mu_suhu_demam(suhu)
mu_tinggi = mu_suhu_tinggi(suhu)

mu_t_low, mu_t_normal = trombosit_low_normal(trombosit)
mu_l_low, mu_l_normal = leukosit_low_normal(leukosit)

alpha1 = min(mu_tinggi, mu_t_low, mu_l_low)

alpha2 = min(mu_demam, mu_t_low)

z1 = 66.668   # output rule 1
z2 = 90.0     # output rule 2

numerator   = alpha1 * z1 + alpha2 * z2
denominator = alpha1 + alpha2

Z = numerator / denominator if denominator != 0 else 0.0

if 60 <= Z <= 100:
    diagnosa = "Risiko Tinggi"
elif 30 <= Z <= 70:
    diagnosa = "Risiko Sedang"
else:
    diagnosa = "Risiko Rendah"

print(f"Nilai Risiko DBD = {Z}")
print(f"Diagnosa: {diagnosa}")
```

```python
def trimf(x, a, b, c):
    if x <= a or x >= c:
        return 0.0
    elif a < x < b:
        return (x - a) / (b - a)
    elif b <= x < c:
        return (c - x) / (c - b)
    else:  # x == a atau x == c
        return 0.0

x1 = 60
x2 = 26

mu_x1_low  = trimf(x1, 0, 30, 60)
mu_x1_high = trimf(x1, 40, 70, 100)

mu_x2_low  = trimf(x2, 15, 20, 25)
mu_x2_high = trimf(x2, 22, 28, 35)

w1 = mu_x1_low  * mu_x2_low
w2 = mu_x1_low  * mu_x2_high
w3 = mu_x1_high * mu_x2_low
w4 = mu_x1_high * mu_x2_high

wsum = w1 + w2 + w3 + w4
if wsum == 0:
    w1n = w2n = w3n = w4n = 0.0
else:
    w1n, w2n, w3n, w4n = w1/wsum, w2/wsum, w3/wsum, w4/wsum

p1, q1, r1 = 0.0692, 0.0461, 0.00231
p2, q2, r2 = 0.0698, 0.0390, 0.00139
p3, q3, r3 = 0.0585, 0.0183, 0.00084
p4, q4, r4 = 0.0505, 0.0318, 0.00115

f1 = p1 * x1 + q1 * x2 + r1
f2 = p2 * x1 + q2 * x2 + r2
f3 = p3 * x1 + q3 * x2 + r3
f4 = p4 * x1 + q4 * x2 + r4

y = w1n * f1 + w2n * f2 + w3n * f3 + w4n * f4

print(f"Output Produksi Jagung = {y} ton/ha")
```
